```python
import pandas as pd
import os
import sys
from IPython.core.display import Image
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import os
from scipy.signal import find_peaks
```

```python
def basic_graph(data):
    plt.rcParams["figure.figsize"] = (14,5)
    fig,ax = plt.subplots()
    data['Mean(2->1Bytes)'] = np.mean(data['2->1Bytes'])
    data['Median(2->1Bytes)'] = np.median(data['2->1Bytes'])
    x = list(range(len(data)))
    ymean1 = [data['Mean(2->1Bytes)'][0] for i in x]
    ymid1 = [data['Median(2->1Bytes)'][0] for i in x]
    line1 = ax.plot(x,data["2->1Bytes"],label='Data')
    meanline1 = ax.plot(x,ymean1, label='Mean='+str(round(data['Mean(2->1Bytes)'][0])), linestyle='--')
    medianline1 = ax.plot(x,ymid1, label='Median='+str(round(data['Median(2->1Bytes)'][0])), linestyle='--')
    ax.legend()
    ax.set_ylabel("bytes")
    ax.set_title("traffic of data streaming, 2->1Bytes")
    return plt

def basic_table(data):
    return data[['1->2Bytes','2->1Bytes','1->2Pkts','2->1Pkts']].describe()

def spike_graph(data):
    x = data['2->1Bytes']
    peaks, _ = find_peaks(x, height=300000)
    plt.figure(figsize=(14, 5))
    plt.plot(x)
    plt.plot(peaks, x[peaks], "x")
    plt.title("This data has " +str(len(peaks))+" of spikes thats larger than 300000 bytes")
    plt.xlabel('Time')
    plt.ylabel('Bytes')
    return plt
```

## Andrey's Data

```python
temp = pd.read_csv('apristin-twitch-live-20210108-12.csv', index_col=0).reset_index()
print(basic_table(temp))
print(basic_graph(temp))
print(spike_graph(temp))
```
